```python
import pandas as pd
from collections import defaultdict
from operator import itemgetter
```

```python
def Borda(df):
  res=[]
  n_voters = df.shape[1]-1
  n_candidates = len(df['candidate'])
  scores = defaultdict(int)
  for i in range(0, n_candidates):
    for j in range(1, n_voters+1):
      scores[df['candidate'][i]] +=df.iloc[i,j] 
  # Sort candidates by score, descending
  ranking = sorted(scores.items(), key=lambda i: i[1], reverse=True)
  for cand, score in ranking:
    #print(cand, score)
    res.append([cand, score])
    #print('='*20)
  return res

def sum_lists(list1, list2):
  overall=[]
  for i in range(0,len(list1)):
    overall.append([list1[i][0],list1[i][1]+list2[i][1]])
  return overall
```

```python
def as_dataframe(res):
  list1=[]
  list2=[]
  for i in range(0, len(res)):
    list1.append(res[i][0])
    list2.append(res[i][1])
  df = pd.DataFrame(list(zip(list1, list2)),columns =['Methods', 'Votes_sum'])
  df["Ranks"] = df["Votes_sum"].rank(method='dense',ascending=False)
  return df
```
